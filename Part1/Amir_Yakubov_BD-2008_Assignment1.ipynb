```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
from collections import Counter
from scipy.spatial import distance
import math
```

## Task 1.

1, 2.

```python
with open("words.txt", "r") as f:
    data = f.read()
```

3.

```python
tokenization = re.split('[^a-z]', data.lower())
while '' in tokenization:
    tokenization.remove('')
```

4.

```python
unique_words = np.unique(tokenization)
indexing = dict(enumerate(unique_words))
```

```python
sentences = []
with open('words.txt') as file:
    for line in file:
        sentences.append(re.findall(r'[a-z]+',line.lower()))
```

```python
length = len(sentences)
```

5.

```python
matrix=[]
for i in sentences:
    present = dict(Counter(i))
    p_o_n = [present[yes] if yes in present else 0 for no,yes in indexing.items()]
    matrix.append(p_o_n)
print(matrix)
```

```python
df = pd.DataFrame(matrix, columns = unique_words)
df
```

6.

```python
cosine_distance = [[distance.cosine(matrix[0],matrix[i]),i] for i in range(length)]
scd = sorted(cosine_distance)
cd = pd.DataFrame(scd, columns = ['cosine_distance', 'sentences'])
cd[1:3]
```
